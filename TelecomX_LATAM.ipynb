```python
#manipulacion de columnas categoricas
cols = ['Churn', 'customer.gender',
       'customer.Partner', 'customer.Dependents',
       'phone.PhoneService', 'phone.MultipleLines', 'internet.InternetService',
       'internet.OnlineSecurity', 'internet.OnlineBackup',
       'internet.DeviceProtection', 'internet.TechSupport',
       'internet.StreamingTV', 'internet.StreamingMovies', 'account.Contract',
       'account.PaymentMethod']

for col in cols:
    datos_normal[col] = datos_normal[col].str.lower()
    datos_normal[col] = datos_normal[col].str.replace('[^a-zA-Z0-9\-]', ' ', regex=True)
    datos_normal[col] = datos_normal[col].str.replace('(?<!\w)-(?!\w)', '', regex=True)
    datos_normal[col] = datos_normal[col].str.strip()

datos_normal.sample(10)
```

##**DistribuciÃ³n de la variable "churn"**
###Comprender cÃ³mo estÃ¡ distribuida la variable "churn" (BajaDelCliente) entre los clientes.
Utiliza grÃ¡ficos para visualizar la proporciÃ³n de clientes que permanecieron y los que se dieron de baja.

```python
import matplotlib.pyplot as plt

# Calcular los porcentajes
datos_baja = datos_normal['BajaDelCliente'].value_counts(normalize=True).to_frame()
datos_baja.columns = ['Porcentaje']

# Crear la grÃ¡fica
ax = datos_baja.plot(kind='bar', figsize=(7, 5), color='green')
ax.set_xlabel('Baja del Cliente')
ax.set_ylabel('Porcentaje')
ax.set_title('Porcentaje de Baja del Cliente')

# Cambiar etiquetas del eje X
nuevas_etiquetas = ['Activo', 'Dado de Baja']
ax.set_xticklabels(nuevas_etiquetas, rotation=0)

# Agregar etiquetas con los porcentajes encima de cada barra
for i, valor in enumerate(datos_baja['Porcentaje']):
    ax.text(i, valor + 0.01, f'{valor:.2%}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()

plt.savefig("porcentaje_baja_cliente.png", dpi=300, transparent=True)

plt.show()

```

##**Recuento de Bajas por variables categÃ³ricas**
###Explorar cÃ³mo se distribuye la Baja de clientes segÃºn variables categÃ³ricas, como gÃ©nero, tipo de contrato, mÃ©todo de pago, entre otras.
Este anÃ¡lisis puede revelar patrones interesantes, por ejemplo, si los clientes de ciertos perfiles tienen una mayor tendencia a cancelar el servicio, lo que ayudarÃ¡ a orientar acciones estratÃ©gicas.


```python
col_categ = ['cliente.Genero', 'telefono.MultiplesLineas',
       'internet.ServicioDeInternet', 'internet.SeguridadOnline',
       'internet.BackupOnline', 'internet.SeguridadDispositivo',
       'internet.SoporteTecnico', 'internet.TVenLinea',
       'internet.PeliculasEnLinea', 'cuenta.TipoContrato',
       'cuenta.FacturaElectronica', 'cuenta.MetodoDePago']


import plotly.express as px

# Diccionario para mapear valores originales a etiquetas personalizadas
map_leyenda = {1: 'Se dio de baja', 0: 'Activo'}

for col in col_categ:
    # Crear una serie mapeada para usar en color
    color_series = datos_normal['BajaDelCliente'].map(map_leyenda)

    fig = px.histogram(
        datos_normal,
        x=col,
        color=color_series,
        barmode='group',
        text_auto=True,
        title=f'DistribuciÃ³n de bajas segÃºn la variable categÃ³rica: {col}',
        labels={'color': 'Estado del Cliente'},
        width=1000,
        height=600
    )

    fig.write_html(f"bajas_segun_VarCat_{col}.html")
    fig.show()
```

##**Conteo de Bajas por variables numÃ©ricas**
###Explora cÃ³mo las variables numÃ©ricas, como "total gastado" o "tiempo de contrato", se distribuyen entre los clientes que cancelaron (Baja) y los que no cancelaron.
